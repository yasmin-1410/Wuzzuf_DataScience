```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time

browser = webdriver.Chrome()
browser.get('https://wuzzuf.net/jobs/egypt')
search = browser.find_element(By.XPATH, "//input[@name='q']")
search.click()
search.send_keys("Software Engineer")
search.send_keys(Keys.ENTER)
time.sleep(3)

base_url = 'https://wuzzuf.net/search/jobs/?a=hpb&q=Software%20Engineer&start='

def extract_jobs():
    job_titles = browser.find_elements(By.XPATH, '//h2[@class="css-m604qf"]')
    company_names = browser.find_elements(By.CLASS_NAME, 'css-17s97q8')
    locations = browser.find_elements(By.CLASS_NAME, 'css-5wys0k')
    date_posted = browser.find_elements(By.XPATH, '//div[@class="css-4c4ojb"]')
    descriptions = browser.find_elements(By.CLASS_NAME, 'css-y4udm8')
    
    jobs = []
    for i in range(min(len(job_titles), len(company_names), len(locations), len(date_posted), len(descriptions))):
        job = {
            'title': job_titles[i].text,
            'company': company_names[i].text,
            'location': locations[i].text,
            'date_posted': date_posted[i].text,
            'descriptions': descriptions[i].text
        }
        jobs.append(job)
    return jobs

all_jobs = []
start = 0

while start <= 5:
    url = base_url + str(start)
    browser.get(url)
    time.sleep(3)
    jobs = extract_jobs()
    all_jobs.extend(jobs)
    start += 1

df = pd.DataFrame(all_jobs)
filename = 'wuzzuf_software_engineer_jobs.csv'
df.to_csv(filename, index=False)
browser.quit()
pd.read_csv('wuzzuf_software_engineer_jobs.csv')
```
